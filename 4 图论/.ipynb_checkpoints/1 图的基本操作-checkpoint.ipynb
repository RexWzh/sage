```python
"1.初始化图"
G1 = graphs.EmptyGraph() #新建空图
G2 = Graph([(1,2),(3,4)]) #通过点对新建
G3 = Graph({1:[1,2],0:(3,4)}) #通过邻接表新建
G4 = graphs.CompleteGraph(6) #新建完全图
G5 = graphs.PathGraph(5) #新建路径图
show(Graph(4)) #直接新建4个点的图
show(G2);show(G3);show(G4);show(G5)
```

```python
"2.添加，删除边"
#添加一边
G1 = Graph() #空图
u,v = 2,1
G1.add_edge(u,v) #不检查是否已有点集
show(G1)

#添加多边
G2 = Graph()
G2.add_edges([(0,1),(2,1),(3,0),(3,1),(3,2)])
show(G2)

#添加圈
G3 = Graph([(0,2)])
G3.add_cycle(list(range(4)))
show(G3)

#添加路径
G4 = Graph([(1,3)])
G4.add_path(list(range(5)))
show(G4)

#delete_edge,delete_edges，与添加类似
```

```python
"3.添加，删除点"
G = Graph([(0,2)])
G.add_vertex('a') #添加一点，可以为任意对象，默认根据图中已有序号添加最小序号顶点
show(G)
G.add_vertices(range(4)) #加入若干点集
show(G)
G.delete_vertex('a') #删除一点
show(G)
G.delete_vertices([1,0]) #删除多点
show(G)
```

```python
"4.无向图转有向图（边转双向边）"
G = Graph([(1,2),(2,0)])
G.to_directed()
```

```python
"5.查看点集，边集"
G = graphs.CompleteGraph(4)
print('G的边集',G.edges(sort=False))
print('G的点集',G.vertices()) #与get_vertices很类似？
print('G的邻接表',G.to_dictionary())
```

```python
"6.图的邻接矩阵"
G = graphs.CompleteGraph(4)
print('G的邻接矩阵',G.adjacency_matrix(),sep='\n')
```

```python
"7.判断图"
G = graphs.CompleteGraph(4)
print('G是否含圈',G.has_loops())
print('G是否含点3',G.has_vertex(3))
print('G是否含边(1,2)',G.has_edge(1,2))
G1 = Graph([(1,2),(0,3)])
G2 = Graph([(1,3)])
print('G2是否为G1子图',G2.is_subgraph(G1))
print('G1是否连通',G1.is_connected())
```

```python
"8.点操作"
G = graphs.CompleteGraph(4)
print('0邻接的点集',G.neighbors(0))
show(G)

G.merge_vertices([1,2])
print('将G的点1,2合并')
show(G)

print('从顶点1出去的边',G.edges_incident(1))

print('顶点1的度数',G.degree(1))
```
