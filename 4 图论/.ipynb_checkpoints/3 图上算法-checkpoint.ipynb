```python
import matplotlib
matplotlib.rcParams['font.sans-serif'] = ['SimHei'] #设置中文编码
from pprint import pprint
```

```python
#label作为权值
weight = lambda e:e[-1]
```

```python
"""最小生成树-Prim，Kruskal
目的：各点连接的最小成本
应用：公路网，通讯网，最小连通成本
"""
G = Graph([(1,2,1),(1,3,2),(2,3,2),(2,4,4),(2,5,3),(3,4,4),(3,5,4),(4,5,2)])
G.show(title="图 G")

print('Kruskal算法求最小生成树')
mst = G.min_spanning_tree(algorithm='Kruskal_Boost',weight_function=weight)
mst = Graph(mst)
print('G的权值总和',sum(e[-1] for e in mst.edges()))
mst.show(title='G的最小生成树')

print('Prim算法求最小生成树')
mst = G.min_spanning_tree(algorithm='Prim_Boost',weight_function=weight)
mst = Graph(mst)
print('G的权值总和',sum(e[-1] for e in mst.edges()))
mst.show(title='G的最小生成树')
```

```python
"""最短路-Dijkstra，Floyd
目的：两点间最短路
应用：运输
"""
G = DiGraph([(1,2,5),(1,4,3),(1,6,4),
           (2,3,3),(2,4,2),
           (3,6,2),
           (4,3,7),(4,5,5),
           (5,2,4),(5,3,1),
           (6,5,6)],weighted=weight)
G.show(title = "图 G")

sp_D = G.shortest_paths(1,algorithm="Dijkstra_Boost",by_weight=True) #1对多
print('从端点1出发，到各点的最短路')
pprint(sp_D)
print('从端点1出发，到各点的最短距离')
dist_D = G.shortest_path_lengths(1,algorithm='Dijkstra_Boost',by_weight=True) #1对多
print(dist_D,'\n')

dist,pred = G.shortest_path_all_pairs(algorithm="Floyd-Warshall-Python",by_weight=True)
print('求任意两点间的最短路')
print('任意两点的距离')
pprint(dist)
print('任意两点的路径（存储前一节点）')
pprint(pred)
```

```python
"""最大流问题
目的：有向图每条弧流量有限，求最大流
应用：交通网的运输能力"""
G = DiGraph([(1,2,8),(1,3,2),
           (2,4,5),(2,3,6),
           (3,5,8),
           (4,3,5),(4,6,9),
           (5,2,3),(5,4,6),(5,6,7)])
G.show(title="图 G")
cardinal,flow_graph = G.flow(1,6,value_only=False,algorithm="FF") #使用 Ford-Fulkerson 算法
print('最大流',cardinal)
print('最大流的路径及流量')
print(flow_graph.edges())
flow_graph.show(title='最大流图像')
```
