```python
"""不带权的无向图"""
#通过邻接表新建
d = {0:[1,2,3],2:[1,0]}
Graph(d).show() 

#邻接表新建，使用字典给边加标签
d = {0:{1:'a',2:'b',3:'c'},2:{1:'d',0:'e'}}
Graph(d).show() 

#新建n个点的图
Graph(5).show()

#通过邻接矩阵新建
a = matrix([[1,1,1],
           [1,1,0],
           [1,0,0]])
Graph(a,format="adjacency_matrix").show()

#通过关联矩阵新建（略）
```

```python
"""不带权的有向图"""
#通过邻接表新建
d = {0:[1,2,3],2:[1,0]}
DiGraph(d).show() 

#邻接表新建，使用字典给边加标签
d = {0:{1:'a',2:'b',3:'c'},2:{1:'d',0:'e'}}
DiGraph(d).show() 

#新建n个点的图
DiGraph(5).show()

#通过邻接矩阵新建
a = matrix([[0,0,1],
           [1,0,0],
           [0,0,0]])
DiGraph(a,format="adjacency_matrix").show()

#通过关联矩阵新建（略）
```
