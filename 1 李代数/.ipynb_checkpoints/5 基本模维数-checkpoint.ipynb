```python
# 方法一 直接使用维数的内积计算公式
Lie = [["A",range(1,9)],["B",range(2,9)],["C",range(3,9)],["D",range(4,9)],
       ["E",range(6,9)],["F",[4]],["G",[2]]]
res = [] #计算结果
for s,l in Lie:
    for n in l:
        res.append([s+str(n)]) #添加一个结果
        R = RootSystem([s,n])
        V = R.ambient_space()
        positive_roots = [root.to_vector() for root in V.positive_roots()] #正根系
        fundamental_weights = [weight.to_vector() for weight in V.fundamental_weights()] #基本支配权
        rho = sum(fundamental_weights) #rho值
        for w in fundamental_weights: #遍历支配权
            dim = 1
            for a in positive_roots:
                dim *= (w+rho)*a/(rho*a) #乘上d_alpha值
            res[-1].append(dim)
```
