```python
"""验证W^{\phi}翻转==W^{s_i}最长元"""
Lie = [["A",range(1,9)],["D",range(4,9)],["E",range(6,9)],
       ["B",range(2,9)],["C",range(3,9)],["F",[4]],["G",[2]]]
for s,List in Lie:
    res = []
    res2 = []
    for l in List:
        print(s+'%d'%l,end='\t')
        cartan_type = CartanType([s,l])
        csa = CSA(cartan_type)
        csa.standard_form.display = 'vector'
        c = csa.alcove_center() # alcove加权中心
        res.append([csa.ascending_chain(c-t_x) for t_x in csa.fundamental_weights_check()])
        res2.append([csa.ascending_chain(-t_x) for t_x in csa.fundamental_weights_check()])
        print([a==b for a,b in zip(res[-1],res2[-1])])
    print()
```

```python
def prod(mats):
    '''矩阵列表乘积'''
    assert len(mats),'输入不能为空'
    res = mats[0]
    for mat in mats[1:]:
        res = res*mat
    return res
def order(mat):
    '''计算矩阵的阶'''
    assert mat.is_square(),'输入必须为方阵'
    n = 1
    new_mat = mat
    while not new_mat.is_one():
        n += 1
        new_mat *= mat
    return n
```

```python
"""最长元取逆性质，以及阶数"""
cartan_type = CartanType("A10")
csa = CSA(cartan_type)
c = csa.alcove_center() # alcove 中心元
simple_refs = csa.simple_reflections()
res = []
for tx in csa.fundamental_weights_check():
    seq = csa.ascending_chain(-tx)
    mats = [simple_refs[i-1] for i in seq] # 作用矩阵
    res.append(prod(mats))
print(cartan_type)
print([res.index(i^-1) for i in res])
print([order(i) for i in res])
```

```python
"""最长元生成群"""
cartan_type = CartanType("D4")
csa = CSA(cartan_type)
c = csa.alcove_center() # alcove 中心元
simple_refs = csa.simple_reflections()
res = []
for tx in csa.fundamental_weights_check():
    seq = csa.ascending_chain(-tx)
    mats = [simple_refs[i-1] for i in seq] # 作用矩阵
    res.append(prod(mats))
print(cartan_type)
print([order(i) for i in res])
mg1,mg2 = MatrixGroup(res),MatrixGroup(simple_refs)
print(mg1.cardinality(),mg1.structure_description())
print(mg2.cardinality(),mg2.structure_description())
```

```python
"""绘制有限群部分构成的框线图"""
s = "B"
cartan_type = CartanType([s,3])
# csa = CSA(cartan_type)
V = RootSystem([s,3,1]).ambient_space()
W = V.weyl_group()

elements = []
i = 0
while len(elements)<WeylGroup(cartan_type).cardinality(): # A 24 B 48
    e = W[i]
    if 0 not in e.reduced_word():
        elements.append(e)
        print(len(elements),i,end='\t')
    i += 1

p = V.plot_alcoves(alcoves=elements, wireframe=True,alcove_labels=True)
p += V.plot_fundamental_weights()
p.show(frame=False)
```
