```python
'''可选参数
1.roots 选择显示哪些根系
    simple（默认）
    all
    列表
    False（不显示）
2.coroots 余根，同1，默认False
3.fundamental_weights 基本支配权
    None（不显示，默认）
    布尔值
4.reflection_hyperplanes 反射平面，同1
5.fundamental_chamber 基本室（仿射型为alcove）
    None（不显示，默认）
    classical（经典类）
    布尔值
6.alcoves 
    布尔值（默认True）
    列表
7.alcove_labels 标签
8.bounding_box 边界（默认3）
9.alcove_walk
10.projection 设置投射
11.color 设置颜色
    按Dynkin图分配颜色：黑0蓝1红2绿3
12.labels 是否显示标签（布尔值）
'''
```

```python
'''6.只看选定的部分'''
L = RootSystem(["A",3,1]).ambient_space()
W = L.weyl_group()
L.plot(reflection_hyperplanes=False, alcoves=[W.one()], bounding_box=2,color=lambda i:"black")
# L.plot(reflection_hyperplanes=False, alcoves=W.gens(), bounding_box=2)
```

```python
"""7.只画框架"""
L = RootSystem(["A",3,1]).ambient_space()
W = L.weyl_group()
alcoves = [~w for d in range(5) for w in W.affine_grassmannian_elements_of_given_length(d)]
p = L.plot_fundamental_chamber("classical")
p += L.plot_alcoves(alcoves=alcoves[:-3], wireframe=True,color=lambda i:"black")
p += L.plot_fundamental_weights()
p.show(frame=False)
```

```python
'''8.二维走动'''
L = RootSystem(["A",2,1]).ambient_space()
w1 = [0,2,1]
L.plot(alcove_walk=w1, bounding_box=3).show(figsize=12)
```
