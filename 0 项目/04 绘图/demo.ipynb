```python
# 三维杨辉三角
def str2weyl(string,gens):
    '''字符串转weyl群元素'''
    if not len(string):return gens[0]*gens[0]^-1
    if len(string)==1:return gens[int(string)]
    elements = [gens[int(i)] for i in string]
    return reduce(lambda x,y:x*y,elements)

myhash = lambda mat:hash(tuple(i for line in Matrix(mat) for i in line))

def next_levels(list0,gens):
    '''由生成下一阶对应的元素'''
    list1,list2,list3 = [set() for i in range(3)]
    ### 处理字符串 ###
    for string in list0:
        last = string[-1] # 最后一个字符
        list1 = list1.union({string+i for i in "0123" if i!=last}) # 新一层解
        list2 = list2.union({s+i for s in list1 for i in "0123" if i not in s[-2:]})
        list3 = list3.union({s+i for s in list2 for i in "0123" if i not in s[-3:]})
    ### 处理weyl群元素重复 ###
    mats = set()
    res = []
    for List in [list1,list2,list3]:
        res.append([])
        for s in List:
            ms = myhash(str2weyl(s,gens)) # 转矩阵哈希值
            if ms not in mats:
                res[-1].append(s)
                mats.add(ms)
    return res

def words_level(n,gens):
    """返回前n层解的字符串,基本alcove为第1阶"""
    assert n>0,'阶数必须为正'
    if n==1:return [[""]] # 第一层
    tags = [[""],["0"]]
    while n-1:
        tags.extend(next_levels(tags[-1],gens))
        n -= 1
    return tags[:-1]

def triangle_length_d3(n,W):
    '''三维空间上的杨辉三角，初始长为1'''
    assert n>0,'长度不能为负'
    
    res = [[W[0]]] # 第一层解
    if n==1:return res
    
    # 生成字符串信息
    gens = W.gens() # 生成元
    tags = words_level(n,gens)
    for level in tags[1:]:
        res.append([str2weyl(i,gens) for i in level])
    return res
```

```python
s = "A"
V = RootSystem([s,3,1]).ambient_space()
W = V.weyl_group()
gens = W.gens()

elements = [str2weyl(i,gens) for i in ["","0","01","03","013","031","012","032","0310"]]
elements = [str2weyl(i,gens) for i in ["","0132","012","032","0310","0310210","0310230","03102132",'012301','012321','0123023',
                                      '032103', '032123', '0321021']]


p = V.plot_alcoves(alcoves=elements, wireframe=True,alcove_labels=True)
# p += V.plot_fundamental_weights()
p += V.plot_fundamental_chamber("classical")
p.show(frame=False)
```
