```python
load('kostant.sage')
load('positive_decomposition.sage')
```

```python
'''kostant_partition_fun 的两种算法'''
res,res2 = [],[]
depth = 10
for s in "BCD":
    for i in range(depth):
        res.append([])
        res2.append([])
        for j in range(i+1):
            res[-1].append(kostant_partition_fun([s,depth],(i,j),method='recursive'))
            res2[-1].append(kostant_partition_fun([s,depth],(i,j),method='constructive'))
    print(s,res==res2)
table(res) # 打印结果
```

```python
'''正根分解的两种算法'''
def list2set(l):
    dic = {}
    for i in l:
        if i not in dic:
            dic.update({i:1})
        else:
            dic[i] += 1
    return set(dic[key]*key for key in dic.keys())
'''检验数值'''
Lie = [["A",range(1,8)],["D",range(4,8)],["B",range(2,8)],["C",range(2,8)]]
for s,List in Lie:
    for l in List:
        print(s+'%d'%l,end='\t')
        cartan_type = CartanType([s,l])
        positive_roots = list(RootSystem(cartan_type).root_space().positive_roots())
        # BFS 搜索算法
        M = root_system_matrix(cartan_type)
        v = highest_root(cartan_type)
        sols = positive_decomposition(M,v)
        sols = [dec for line in sols for dec in line]
        res = [set(dec2root(dec,positive_roots)) for dec in sols]
        # 构造算法
        sols2 = positive_decomposition_recursive(cartan_type)
        res2 = [list2set(line2root(line,positive_roots)) for line in sols2]
        
        print(len(res)==len(res2),end='\t') # 比较阶数
        for i in res: # 比较内容
            if i not in res2:
                print(False,i)
                break
        else:
            print(True)
    print()
```
