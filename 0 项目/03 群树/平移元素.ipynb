```python
# A-E族wx计算
Lie = [["A",range(1,6),lambda i:i],
       ["B",range(3,6),lambda i:2*i-1],
       ["C",range(2,6),lambda i:(i*i+i)/2],
       ["D",range(4,7),lambda i:(i*i-i)/2],
       ["E",range(6,7),lambda i:16]]
for s,ran,f in Lie:
    for l in ran:
        t= time.time()
        print(s+'%d'%l)
        n = l + 1 # 矩阵列数
        A = CartanMatrix([s,l,1])
        reflections = [] # 单反射
        for i in range(1,n): # W0
            ref  = matrix.identity(n) # n阶单位阵（单反射）
            ref[i] = [-1 if i==j else -A[j,i] for j in range(n)]
            reflections.append(ref)
        W = GroupTree(reflections,index=tuple('%d'%i for i in range(1,n)),length=f(l),elements=False)
#         assert W.card==MatrixGroup(reflections).cardinality(),'群阶数不对'
        T = transform_mat([s,l],dx=-1) # 平移矩阵
        Zeta = affine_graph_automorphism([s,l]) # 图自同构
        for z in Zeta.elements:
            hm = W.myhash(T^-1*z)
            if hm in W.expressions:
                print(z,'\n',''.join(W.expressions[hm]),'\t%.3fs\n'%(time.time()-t),sep='')
```

```python
# D族x''的wx计算
s = "D"
for l in range(4,7):
    print(s+'%d'%l)
    t = time.time()
    n = l + 1 # 矩阵列数
    A = CartanMatrix([s,l,1])
    reflections = [] # 单反射
    for i in range(1,n): # W0
        ref  = matrix.identity(n) # n阶单位阵（单反射）
        ref[i] = [-1 if i==j else -A[j,i] for j in range(n)]
        reflections.append(ref)
    W = GroupTree(reflections,index=tuple('%d'%i for i in range(1,n)),length=(l*l+l)/2)
    T = transform_mat([s,l],dx=-2) # 平移矩阵
    Zeta = affine_graph_automorphism([s,l]) # 图自同构
    for z in Zeta.elements:
        hm = W.myhash(T^-1*z)
        if hm in W.expressions:
            print(z,'\n',''.join(W.expressions[hm]),'\t%.3fs\n'%(time.time()-t),sep='')
```
