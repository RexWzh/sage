```python
class Oth(Oth):
    @staticmethod
    def symbols(name,num):
        """"""
        if num==1:
            return (var(name),)
        return var((name+'%d ')*num%tuple(i+1 for i in range(num)[::-1]))
```

```python
L = Oth(3,root_name='alpha')
dec = L.positive_dec(L.positive_roots[-1])
negative_roots = L.negative_roots #
dec_root = []
for line in dec:
    new = []
    for i,element in enumerate(line):
        if element:
            new = new + [negative_roots[i]]*element
    dec_root.append(tuple(new))
print('',len(dec)) #
table([[-i for i in line if i] for line in dec_root]) #
```
